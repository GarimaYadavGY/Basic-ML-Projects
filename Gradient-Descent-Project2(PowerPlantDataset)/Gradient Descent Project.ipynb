```python
def cost(x,y,m):
    n=len(x)
    
    cost1=0
    for i in range(n):
        mx=0
        for j in range(8):
            mx+=m[j]*x[i][j]
        cost1+=((y[i]-mx)**2)
        
    #print("here")
    return (1/n)*cost1
```

```python
def step_gradient(x,y,m,learning_rate):
    m_slope=np.zeros(8)
    p=len(x)
    
    for i in range(p):
        sum=0
        for j in range(8):
            sum+=m[j]*x[i][j]
        sum=y[i]-sum
        
        for j in range(8):
            m_slope[j]+=(-2/p)*(x[i][j])*sum
            
    m=m-learning_rate*m_slope
    
    return m
```
